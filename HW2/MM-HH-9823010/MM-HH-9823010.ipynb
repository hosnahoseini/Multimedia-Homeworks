```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

## 1)

```python
img = Image.open('image.png')
```

```python
def gray_scale(img):
    rgb = np.array(img)
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray_img = 0.2989 * r + 0.5870 * g + 0.1140 * b
    gray_img = gray_img.astype(int)
    return gray_img 
```

```python
gray_img = gray_scale(img)
plt.imshow(gray_img)
```

## 2)

```python
def create_histogram(array):
    occurrence_array = np.bincount(array.flatten(), minlength=256)
    return occurrence_array
histogram_array = create_histogram(gray_img)
```

```python
plt.plot(histogram_array)
```

## 3)

```python
def count_cumulative_array(array):
    c_array = np.cumsum(array)
    return c_array
```

```python
cumulative_array = count_cumulative_array(histogram_array)
```

```python
plt.plot(cumulative_array)
```

## 4)

```python
def create_mapper(img, color_levels, cumulative_array):
    width = gray_img.shape[1]
    height = gray_img.shape[0]
    color_levels = 256
    transform_map = np.round((color_levels - 1) * cumulative_array / (height * width))
    transformed_image = [transform_map[pixel] for pixel in list(gray_img.flatten())]
    transformed_image = np.reshape(np.asarray(transformed_image), gray_img.shape)
    return transformed_image
```

```python
output_image_array = create_mapper(gray_img, 256, cumulative_array)
output_image_array = np.uint8(output_image_array)
output_image_array
```

## 5)
